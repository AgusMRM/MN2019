```python
import numpy as np
import matplotlib.pylab as plt
```

```python
#rg=np.loadtxt('orbita4orden.dat')
dkd_k=np.loadtxt('orbitaDKD_kep.dat')
kdk_k=np.loadtxt('orbitaKDK_kep.dat')
dkd_p=np.loadtxt('orbitaDKD_plu.dat')
kdk_p=np.loadtxt('orbitaKDK_plu.dat')
dkd_i=np.loadtxt('orbitaDKD_iso.dat')
kdk_i=np.loadtxt('orbitaKDK_iso.dat')
```

```python
G  = 43.03e3/1e10
M0 = 9e10
T0 = .25
r0 = 8.101
v0 = 2*np.pi*r0/T0
```

```python
#fig=plt.figure(1,figsize=(8,8))
fig=plt.figure(figsize=(8,8))
fig.subplots_adjust(bottom=0.08,top=0.98,left=.15,right=.7)

ax1=fig.add_subplot(211)
ax1.plot(kdk_i[:,0],kdk_i[:,1],color='red',label='isocrono')
ax1.plot(kdk_p[:,0],kdk_p[:,1],color='green',label='plummer')
ax1.plot(kdk_k[:,0],kdk_k[:,1],color='blue',label='kepler')
plt.title('KDK')
plt.legend()
plt.xlabel('x')
plt.ylabel('y')
#ax2=fig.add_subplot(212)
#ax2.plot(kdk_p[:,0],kdk_p[:,1])

ax2=fig.add_subplot(212)
ax2.plot(dkd_i[:,0],dkd_i[:,1],color='red',label='isocrono')
ax2.plot(dkd_p[:,0],dkd_p[:,1],color='green',label='plummer')
ax2.plot(dkd_k[:,0],dkd_k[:,1],color='blue',label='kepler')
plt.title('DKD')
plt.legend()
plt.xlabel('x')
plt.ylabel('y')

```

```python
E0 = -G*M0/r0 + .5*v0**2

fig=plt.figure(figsize=(8,8))
fig.subplots_adjust(bottom=0.08,top=0.98,left=.15,right=.7)

ax1=fig.add_subplot(211)
ax1.plot(dkd_k[:,3],(dkd_k[:,2]-E0)/E0,color='blue',label='kepler')
ax1.plot(dkd_i[:,3],(dkd_i[:,2]-E0)/E0,color='red',label='isocrono')
ax1.plot(dkd_p[:,3],(dkd_p[:,2]-E0)/E0,color='green',label='plummer')
ax1.legend()
plt.title('DKD')
plt.xlabel('pasos')
plt.ylabel('energia relativa')

ax2=fig.add_subplot(212)
ax2.plot(kdk_k[:,3],(kdk_k[:,2]-E0)/E0,color='blue',label='kepler')
ax2.plot(kdk_i[:,3],(kdk_i[:,2]-E0)/E0,color='red',label='isocrono')
ax2.plot(kdk_p[:,3],(kdk_p[:,2]-E0)/E0,color='green',label='plummer')
ax2.legend()
plt.title('KDK')
plt.xlabel('pasos')
plt.ylabel('energia relativa')
```
