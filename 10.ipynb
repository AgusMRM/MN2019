```python
import numpy as np
import matplotlib.pylab as plt
```

```python
cic=np.loadtxt('CIC.dat')
ngp=np.loadtxt('NGP.dat')
tsc=np.loadtxt('TSC.dat')
```

```python
x=np.arange(0,10)
plt.plot(x,cic[:,0],color='blue',label='CIC')
plt.plot(x,ngp[:,1],color='red',label='NGP')
plt.plot(x,tsc[:,2],color='green',label='TSC')
plt.legend()
```
