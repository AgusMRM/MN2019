```python
import numpy as np
import matplotlib.pylab as plt
```

```python
#Datos correspondientes a un tiempo virial
t0=np.loadtxt('particulas.dat')
t=np.loadtxt('nbody_out.dat')
e=np.loadtxt('energy.dat')
m=np.loadtxt('momentoangular.dat')
```

```python
plt.plot(t0[:,0],t0[:,1],linestyle='none',marker='.',color='blue',label='iniciales')
plt.plot(t[:,0],t[:,1],linestyle='none',marker='.',color='red',label='finales')
plt.legend()
```

```python
E0=-20799774.0
K=e[:,0]
U=e[:,1]
t=e[:,2]
plt.plot(t,(K+U-E0)/E0,label='error relativo',color='darkblue')
plt.legend()
plt.title('ENERGIA')
```

```python
Lx0=-152924.26706886292
Ly0=-66778.297904968262 
Lz0=85868.797470092773 
Lx=m[:,0]
Ly=m[:,1]
Lz=m[:,2]

L0=np.sqrt(Lx0**2+Ly0**2+Lz0**2)
L=np.sqrt(Lx**2+Ly**2+Lz**2)

plt.plot(t,(Lx-Lx0)/Lx0,color='red',label='Lx',linewidth=.5)
plt.plot(t,(Ly-Ly0)/Ly0,color='blue',label='Ly',linewidth=.5)
plt.plot(t,(Lz-Lz0)/Lz0,color='green',label='Lz',linewidth=.5)
plt.plot(t,(L-L0)/L0,color='black',label='L total',linewidth=2)
plt.legend()
plt.title('Angular Momentum (Relative Error)')
```
