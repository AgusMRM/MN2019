```python
import numpy as np
import matplotlib.pylab as plt
import scipy.stats as sp
```

```python
a=0.413848937
b=1.117587E-8
c=-4.5969497E-2
x=[-3.,-1.,1.,3.]
y0=1.2341E-4
```

```python
def poli(j):
    p=y0 + a*(j-x[0])+b*(j-x[0])*(j-x[1])+c*(j-x[0])*(j-x[1])*(j-x[2])
    return p
```

```python
y=[]
for i in np.arange(-4,4,.1):
    y.append(poli(i))
```

```python
plt.plot(np.arange(-4,4,.1),y)
```
