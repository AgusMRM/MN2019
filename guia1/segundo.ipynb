```python
import numpy as np
import matplotlib.pylab as plt
import scipy.stats as sp
from math import acos
```

```python
Norm=[]
pi=acos(-1.)
A=np.sqrt(2*pi)
for i in np.arange(-3,3,.1):
    Norm.append(A*sp.norm.pdf(i,0,1))
```

```python
x=[-3.,0.,3.]
y=[np.exp(-x[0]**2.),np.exp(-x[1]**2.),np.exp(-(x[2]**2.))]
```

```python
f1x0x1=(y[1]-y[0])/(x[1]-x[0])
f1x1x2=(y[2]-y[1])/(x[2]-x[1])
f2x0x1x2=(f1x1x2-f1x0x1)/(x[2]-x[0])
```

```python
def p2(x):
    p1=y[0]+f1x0x1*(x+3.)
    p2=p1+f2x0x1x2*(x+3.)*x
    return p2
```

```python
y2=[]
x2=[]
for i in np.arange(-3.,3,.1):
    y2.append(p2(i))
    x2.append(i)
```

```python
plt.plot(x2,y2,linestyle='none',marker='.')
```
