```python
import numpy as np
import matplotlib.pylab as plt
```

```python
#abrir archivo 13.f90, el parametro vec es el n-esimo vecino a calcular (aca esta como vec=5), 
#np son las particulas, y cell la cantidad de celdas
```

```python
v=np.loadtxt('vecinas.dat') # -> este no sirve para nada.. 
p=np.loadtxt('pruebas.dat')  # -> contiene todas las particulas, indices, linkeo, box al que pertenecen (x,y)
d=np.loadtxt('distancias.dat') # -> contiene indice de cada particula, y la distancia a la n-esima vecina
```

```python
# EJ, queremos la distancia para la i-esima particula, por ejemplo:
i=34
cir1 = plt.Circle((p[i,0],p[i,1]),d[i,1],color='yellow',alpha=.4)

j= 5
cir2 = plt.Circle((p[j,0],p[j,1]),d[j,1],color='yellow',alpha=.4)

k=99
cir3 = plt.Circle((p[k,0],p[k,1]),d[k,1],color='yellow',alpha=.4)
```

```python
plt.plot(p[:,0],p[:,1],linestyle='none',marker='.')
plt.title('particulas')
plt.plot(p[i,0],p[i,1],marker='*',color='red')
plt.plot(p[j,0],p[j,1],marker='*',color='red')
plt.plot(p[k,0],p[k,1],marker='*',color='red')
fig = plt.gcf() 
fig.gca().add_artist(cir1)
fig.gca().add_artist(cir2)             #-> circulos amarillos centrados en la particula y de radio 
fig.gca().add_artist(cir3)            # la distancia a la n-esima mas cercana
```
